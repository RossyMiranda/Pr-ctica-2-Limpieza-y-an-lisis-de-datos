### 4.2 Normalidad y Homogeneidad de la Varianza de los Datos

#### 4.2.1 Normalidad

Para la comprobación de que los valores que toman nuestras variables cuantitativas provienen de una población distribuida normalmente, utilizaremos la prueba de normalidad de Shapiro_Wilk. Siendo la hipótesis nula que la población está distribuida normalmente, si el p-valor es menor a alfa= 0,05 (nivel de significancia) entonces la hipótesis nula es rechazada (se concluye que los datos no vienen de una distribución normal). Si el p-valor es mayor a alfa= 0,05, se concluye que no se puede rechazar dicha hipótesis.

```python
from statsmodels.graphics.gofplots import qqplot
from scipy.stats import shapiro

fig = plt.figure(figsize=(15,15))
fig.subplots_adjust(hspace=0.4, wspace=0.4)
i=1
for variable in list(X):
    ax = fig.add_subplot(3, 4, i)
    ax.set_title(variable)
    qqplot(X[variable], line='s',ax=ax) # q-q plot
    stat, p = shapiro(X[variable])    
    ax.set_xlabel('Statistics=%.2f, p=%.2E' % (stat, p))
    i=i+1
plt.show()
```

<div style="background-color: #FEF5E7; border-color: #7C9DBF; border-left: 5px solid #7C9DBF; padding: 0.5em;">
<strong> Respuesta:</strong> 
    
Analizando los p-valor de todas las variables aplicando el test de Shapiro_Wilk se observa que se obtienen 
valores inferiores a alfa= 0,05 (nivel de significancia) entonces la hipótesis nula es rechazada 
por lo que se concluye que los datos no vienen de una distribución normal.

</div>


#### 4.2.2 Homogeneidad de la Varianza de los Datos

Se realizará el test de homogeneidad de la varianza de todas las variables dependientes  con relación a la calidad del vino. Para esto utilizatremos la prueba de Levene, como las variables no siguen una distribución normal se utilizará la mediana como métrica de tendencia central, con esto, con un p-valor superior a 0,05, aceptamos la hipótesis de que las varianzas de las muestras son homogéneas.

```python
for variable in list(X):
    statistic,pvalue = stats.levene(X.loc[baja_calidad.index,variable], X.loc[alta_calidad.index,variable], center='median')
    print(variable+' : Statistics=%.2f, p-value=%.2f' % (statistic,pvalue))
```

- residual sugar: p-value=0.41
